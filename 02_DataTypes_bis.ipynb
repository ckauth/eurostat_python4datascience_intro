Python also knows the operators & and | that are similar to 'and' and 'or'

Be careful, they are not the same!

- & and | are bitwise operators
- 'and' and 'or' are logical operators.

Using bitwise operations for vector comparison with objects like lists might lead to unexpected results. We will encounter lists later in this tutorial. We recommend to use 'and' and 'or' by default.

```python
a, b = True, False
print(a & b)
print(a | b)
print((a | b) & (a | b))
print((a | b) | (b | b))
```

Numbers and strings can be converted to booleans. If they are not 0 or not empty, they will evaluate to true.

```python
print(bool(3.5))
print(bool(0))
print(bool("hello"))
print(bool(""))
print(bool([1, 2, 3]))  # see next section for lists
print(bool([]))  # an empty list
```

## Sets

-  A set is a list with no duplicate elements, so only **unique elements**.
-  If an element already exists in the set it is not added.
-  The set is unordered, so subsetting by index is useless.  
-  A set is created with the function `set()`.
-  For sets there are useful operations such as intersection and difference.

```python
x = set(["a", "b", "c"])
type(x)
```

```python
x = {"a", "b", "c"}
type(x)
```

```python
# Create a set
x = set(["a", "b", "c"])

# Add an element to a set
x.add("d")
print(x)

# Add multiple elements to a set with the update() function
x.update(["4", "2"])
print(x)

# Remove an element from the set (remove the element by name, not by index)
x.discard("b")
print(x)
```

```python
dir(set)
```

```python
"my name is Christian and Christian is my name".split()
```

```python
# A set consists of unique elements, so duplicates are removed automatically.
print(set("my name is Christian and Christian is my name".split()))
```

```python
# Create sets
a = set(["Jake", "John", "Eric"])
b = set(["John", "Jill"])

# The intersection of two sets:
print(a.intersection(b))

# The difference between two sets:
print(a.difference(b))
print(a-b)
```
